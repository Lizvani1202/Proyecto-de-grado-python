```python
import PySimpleGUI as sg
import cv2
import numpy as np
import sys
from sys import exit as exit
from datetime import datetime
import requests
import pandas as pd
from tkinter import messagebox
import csv
import time
import asyncio
import aiohttp
import getpass
import base64

import nest_asyncio
nest_asyncio.apply()

import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)

programm_run = True

foto = None
writer = None

lista_placas = []
fechas = []
puntos_registrados = []
file = 'save.csv'  # nombre del archivo de datos previamente creado
max_num_plate = 10
imagen = None

backend_url = 'http://localhost:8090/api'
isAuthenticated = False
cred_base64 = None
current_checkpoint = None


# funcion para leer la placa
async def leer_placa(img):
    regions = ['gb', 'it']
    with open(img, 'rb') as fp:
        # se pide la consulta al servidor
        response = requests.post(
            'https://api.platerecognizer.com/v1/plate-reader/',
            data=dict(regions=regions),
            files=dict(upload=fp),
            headers={'Authorization': 'Token '})
    return response.json()


async def registrar_placa_en_sistema(numero_placa):
    try:
        if not current_checkpoint:
            raise Exception("No se ha seleccionado un checkpoint")
        data = {
            "carNumberPlate": numero_placa,
            "checkpointExitName": current_checkpoint
        }
        
        headers = {
            'Authorization': 'Basic ' + cred_base64,
            'Content-Type': 'application/json'
        }
        
        async with aiohttp.ClientSession() as session:
            async with session.post(backend_url + '/drivers/add-report-penalty', json=data, headers=headers) as resp:
                if resp.status == 200 or resp.status == 201:
                    response = await resp.json()
                    if 'carFeatures' in response:
                        if 'numberPlate' in response['carFeatures']:
                            print("Placa registrada en el sistema:",
                                  response['carFeatures']['numberPlate'])
                    return response
                else:
                    print("Error al registrar placa en el sistema:", resp.status)
                    print(await resp.text())
                    return None
    except Exception as e:
        print("Error al registrar placa en el sistema:", e)
        return None


async def validar_placa(data, lista_placas, file, writer, max_num_plates, fechas):
    # siempre que la imagen sea una placa
    if data is not None and 'results' in data and data['results'] != []:
        # validar que no haya placas repetidas
        for result in data['results']:
            license_plate = result['plate'].upper()
        if len(lista_placas) > 1:  # para comenzar a verificar a partir de 1 placa
            if result['plate'] in lista_placas:
                validar = 1  # filtra placas repetidas
            else:
                lista_placas.append(license_plate)
                response = await registrar_placa_en_sistema(license_plate)
                validar = 0  # asigna cero si no esta repetida
        else:  # se agrega la primera placa
            lista_placas.append(license_plate)
            response = await registrar_placa_en_sistema(license_plate)
            validar = 0
        # verificar que haya menos de maximo numero de carros
        if len(lista_placas) > max_num_plates:
            validar = 2
            lista_placas.pop()  # elimina la ultima placa de la lista
        # se verifica si hay espacio en el parking o si esta repetida la placa antes de guardar
        guardar_placa(data, file, writer, validar)
        # Mensaje para saber que se leyo correctamente la placa
        # messagebox.showinfo(message='PLACA leída "CORRECTAMENTE"')
        # Se lee la fecha y la hora
        date_ = datetime.today().strftime('%x %X')  # leer le fecha y hora
        fechas.append(date_.split(' '))
        puntos_registrados.append(current_checkpoint)
        # date,time_=date_[0], date_[1]

    # else: # cuando la imagen no tenga una placa
    #    messagebox.showerror("ERROR", 'La imagen NO FUE RECONOCIDA')


# funcion para ir guardando en el archivo cada placa leida
def guardar_placa(data, file, writer, validar):
    for result in data['results']:
        if validar == 0:
            writer.writerow(dict(date=datetime.today().strftime('%x %X'),
                                 license_plate=result['plate'],
                                 score=result['score'],
                                 dscore=result['dscore'],
                                 ))
        # if validar==1:
        #    messagebox.showerror("ERROR", 'PLACA REPETIDA')
        # if validar==2:
        #    messagebox.showerror("ERROR", 'MAS de ' + str(max_num_plate)+ ' AUTOS')


def tablero(frame, lista_placas, fechas, puntos_registrados):
    global current_checkpoint
    pizarra = np.zeros((frame.shape[0], 500, 3), np.uint8)
    # titulo
    cv2.putText(pizarra, "Plates readed :", (10, 20),
                cv2.FONT_HERSHEY_COMPLEX, .6, (255, 0, 255), 1, cv2.LINE_AA)
    # imprimir placas en el tablero
    y = 25
    for placa, fecha, punto in zip(lista_placas, fechas, puntos_registrados):
        y += 16
        cv2.putText(pizarra, placa, (10, y),
                    cv2.FONT_HERSHEY_COMPLEX, .5, (255, 254, 255), 1, cv2.LINE_AA)
        cv2.putText(
            pizarra, fecha[0], (90, y), cv2.FONT_HERSHEY_COMPLEX, .5, (255, 254, 255), 1, cv2.LINE_AA)
        cv2.putText(pizarra, fecha[1], (190, y),
                    cv2.FONT_HERSHEY_COMPLEX, .5, (255, 254, 255), 1, cv2.LINE_AA)
        
        cv2.putText(pizarra, punto, (290, y),
                    cv2.FONT_HERSHEY_COMPLEX, .5, (255, 254, 255), 1, cv2.LINE_AA)
    # unir pizarra con imagen
    combinado = np.hstack((frame, pizarra))
    return combinado


async def verificar_imagen():
    global programm_run
    global foto
    global lista_placas
    global file
    global writer
    global max_num_plate
    global fechas
    while programm_run:
        if foto is not None and current_checkpoint is not None:
            datos = await leer_placa(foto)
            if 'status_code' in datos and datos['status_code'] != 200 and 'detail' in datos:
                print('Error al leer placa: ' + datos['detail'])
            await validar_placa(datos, lista_placas, file, writer, max_num_plate, fechas)
        else:
            if foto is None:
                print("No se pudo procesar la imagen")
            if current_checkpoint is None:
                print("No se ha seleccionado un checkpoint")
        await asyncio.sleep(5)


"""
Demo program that read plate
"""
sg.ChangeLookAndFeel('Black')

combo_layout = sg.Combo([], font=('Arial Bold', 14),  expand_x=True, enable_events=True,  readonly=False, key='-COMBO-')

layout = [
    [sg.Text('PLATE ', size=(40, 1), justification='center', font='Helvetica 20'),
    sg.Column(key='-C-', size=(500, 40), layout=[
        [sg.Text(' '*10),
        sg.Text('Email:'), sg.InputText(key='email', size=(15, 1)),
        sg.Text('Password:'), sg.InputText(key='password', size=(15, 1), password_char='*'),
        sg.Button('OK')]
    ], element_justification='right', expand_x=True),
    sg.Column(key='-C2-', size=(500, 40), visible=False, layout=[
        [combo_layout, sg.Text('', key='email_logged'), sg.Button('Cambiar cuenta')]
    ], element_justification='right', expand_x=True)],
    [sg.Image(filename='', key='image', visible=False), 
     sg.Column(key='-C3-', size=(480, 500), layout=[
        [sg.Text('Entre con su cuenta para iniciar el reconocimiento de placas')]
    ], element_justification='center', expand_x=True, visible=True)],
    [sg.Button('Read plate', size=(10, 1), font='Helvetica 14'),
    sg.Button('Map', size=(10, 1), font='Any 14'),
    sg.Button('Exit', size=(10, 1), font='Helvetica 14')]
]

async def main_loop():
    global programm_run
    global foto
    global lista_placas
    global file
    global writer
    global max_num_plate
    global fechas
    global cred_base64
    global isAuthenticated
    global current_checkpoint
    global puntos_registrados
    # define the window layout
    # create the window and show it without the plot
    window = sg.Window('Reconocimiento ', layout, location=(400, 200))
    ######## ciclo para que la ventana se ejecute constantemente #######
    # se bare el archivo para escritura
    with open(file, 'w') as output:
        # se definen los campos del fichero
        fields = ['date', 'license_plate', 'score', 'dscore']
        writer = csv.DictWriter(output, fieldnames=fields)
        writer.writeheader()
        # se hace un ciclo para que la ventana se mantenga constantemente
        # ---===--- Event LOOP Read and display frames, operate the GUI --- #
        cap = cv2.VideoCapture(0)
        recording = True
        while programm_run:
            event, values = window.Read(timeout=20)
            if event == 'Exit' or event == sg.WIN_CLOSED or event == sg.WINDOW_CLOSED:
                programm_run = False
                cap.release()  # libera el objeto de video
                break
            elif event == 'OK':
                email = values['email']
                password = values['password']
                cred_base64 = base64.b64encode(bytes(email + ':' + password, 'utf-8')).decode('utf-8')

                data = {
                    "email": email,
                    "password": password
                }

                headers = {
                    'Content-Type': 'application/x-www-form-urlencoded'
                }

                try:
                    async with aiohttp.ClientSession() as session:
                        async with session.post(backend_url + '/users/login', data=data, headers=headers) as resp:
                            if resp.status == 200:
                                response = await resp.json()
                                if 'accessToken' in response:
                                    isAuthenticated = True
                                    window['-C-'].update(visible=False)
                                    window['-C2-'].update(visible=True)
                                    mensaje = "Cuenta actual: " + email

                                    try:
                                        current_checkpoint = None
                                        async with aiohttp.ClientSession() as session2:
                                            async with session2.get(backend_url + '/location-checkpoints/get-all-location-checkpoints', headers={'Authorization': 'Basic ' + cred_base64}) as resp2:
                                                if resp2.status == 200 or resp2.status == 201:
                                                    response2 = await resp2.json()
                                                    values = []
                                                    for checkpoint in response2:
                                                        values.append(checkpoint['name'])
                                                    content_fit_value = max(values, key=len)
                                                    window['-COMBO-'].update(values=values, size=(len(content_fit_value), len(values)))
                                                    # mensaje_espacios = ' '*(25 - len(mensaje) - len(content_fit_value)*2)
                                                    # window['space'].update(mensaje_espacios)
                                    except Exception as e:
                                        print("Error al obtener checkpoints:", e)

                                    window['email_logged'].update(mensaje)
                                    window['-C3-'].update(visible=False)
                                    window['image'].update(visible=True)

                                else:
                                    print("Error al autenticar usuario")
                            else:
                                print("Error al autenticar usuario:", resp.status)
                except Exception as e:
                    print("Error al autenticar usuario")
                
            elif event == 'Cambiar cuenta':
                isAuthenticated = False
                cred_base64 = None
                window['-C-'].update(visible=True)
                window['-C2-'].update(visible=False)
                window['email'].update('')
                window['password'].update('')
                window['email_logged'].update('')
                window['-C3-'].update(visible=True)
                window['image'].update(visible=False)
                current_checkpoint = None
            # elif event == 'Read plate':
            elif event == '-COMBO-':
                current_checkpoint = values['-COMBO-']
            elif event == 'Map':
                break
            
            
            ret, frame = cap.read()
            if isAuthenticated:
                # Crear tablero de resultados
                frame_tablero = tablero(frame, lista_placas, fechas, puntos_registrados)
                imgbytes = cv2.imencode('.png', frame_tablero)[1].tobytes()  # ditto
                # window.FindElement('image').Update(data=imgbytes)
                # update window with image
                window['image'].update(data=imgbytes)

                foto = "temp.jpg"  # nombre de la imagen temporal a guardar
                # se guarda la imagen capturada por el video
                cv2.imwrite(foto, frame)
            # # se llama a la funcion leer placa
            # data = await leer_placa(foto)
            # validar_placa(data,lista_placas,file,writer,max_num_plate, fechas) #recibe la fecha y hora

            await asyncio.sleep(0)
            # if recording:
            #    ret, frame = cap.read()
            #    #Crear tablero de resultados
            #    frame_tablero=tablero(frame, lista_placas,fechas)
            #    imgbytes=cv2.imencode('.png', frame_tablero)[1].tobytes() #ditto
            #    window.FindElement('image').Update(data=imgbytes)
        output.close()
    window.close()


async def main():
    task_main = asyncio.create_task(main_loop())
    task_verificar_imagen = asyncio.create_task(verificar_imagen())
    await asyncio.gather(task_main, task_verificar_imagen)

await main()
```
